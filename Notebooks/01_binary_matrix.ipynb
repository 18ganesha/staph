# Importing Data
We want to explore what happens if a pair of closely related strains differ at one gene, how likely is it that they'll differ at others. To do this, we've narrowed down the data set to 100 ID'd samples. We are looking at the presence-absence data and the distance matrix.

```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
df_dist = pd.read_csv(r"C:\Users\jojoa\GitHub\staph\data\dist_matrix.csv") 
df_pa = pd.read_csv(r"C:\Users\jojoa\GitHub\staph\data\presence_absence.csv") 
```

```python
df_pa=df_pa.drop(['Non.unique.Gene.name','Annotation','No..isolates','No..sequences',
            'Avg.sequences.per.isolate','Genome.Fragment','Order.within.Fragment',
            'Accessory.Fragment','Accessory.Order.with.Fragment','QC',
            'Min.group.size.nuc','Avg.group.size.nuc','Max.group.size.nuc'],axis=1)
```

```python
df_pa['Gene']=df_pa['Gene'].str.replace('~','_')
```

```python
df_pa=df_pa.fillna(0)
length=len(df_pa.index)
```

# Creating Binary Matrix
Let's replace each string in every index with 1, and the null indecies will remain 0. This gives us a clean and concise binary matrix.

```python
for i in range(0,length-1): #length-1
    if isinstance(df_pa.iloc[i,1], str)==True:
        string=df_pa.iloc[i,1]
        identify = string[0:4]
        df_pa.iloc[i,1]=identify
        mapping = {identify: 1}
        df_pa=df_pa.replace(df_pa.iloc[i,1],1)
    else:
        df_pa.iloc[i,1]=df_pa.iloc[i,1]
```

```python
for i in range(0,length-1): #length-1
    if isinstance(df_pa.iloc[i,2], str)==True:
        string=df_pa.iloc[i,2]
        identify = string[0:4]
        df_pa.iloc[i,2]=identify
        mapping = {identify: 2}
        df_pa=df_pa.replace(df_pa.iloc[i,2],1)
    else:
        df_pa.iloc[i,3]=df_pa.iloc[i,3]
```

```python
for i in range(0,length-1): #length-1
    if isinstance(df_pa.iloc[i,3], str)==True:
        string=df_pa.iloc[i,3]
        identify = string[0:4]
        df_pa.iloc[i,3]=identify
        mapping = {identify: 2}
        df_pa=df_pa.replace(df_pa.iloc[i,3],1)
    else:
        df_pa.iloc[i,3]=df_pa.iloc[i,3]
```
